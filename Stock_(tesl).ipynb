```python
import numpy as np

def calculate_ema(data, period):
  """
  Calculates the Exponential Moving Average (EMA) of a given dataset.

  Args:
    data: A list or array of data points.
    period: The period of the EMA.

  Returns:
    A list or array of EMA values.
  """

  ema = []
  k = 2 / (period + 1)

  for i in range(len(data)):
    if i == 0:
      ema.append(data[i])
    else:
      ema.append((data[i] * k) + (ema[-1] * (1-k)))

  return ema


# Example usage:
data = [10, 12, 15, 13, 16, 18, 17, 20]
period = 3

ema_values = calculate_ema(data, period)

print("Original Data:", data)
print("EMA Values:", ema_values)


```

```python
import matplotlib.pyplot as plt

def plot_ema_chart(data, period):
  """
  Plots a chart of the original data and its Exponential Moving Average (EMA).

  Args:
    data: A list or array of data points.
    period: The period of the EMA.
  """

  # Calculate the EMA values
  ema_values = calculate_ema(data, period)

  # Plot the original data and EMA
  plt.plot(data, label='Original Data')
  plt.plot(ema_values, label='EMA')

  # Set plot title and labels
  plt.title('EMA Chart')
  plt.xlabel('Time')
  plt.ylabel('Value')

  # Display the legend
  plt.legend()

  # Show the plot
  plt.show()


# Example usage:
data = [10, 12, 15, 13, 16, 18, 17, 20]
period = 3

plot_ema_chart(data, period)
```

```python
import matplotlib.pyplot as plt

def plot_fibonacci_chart(data):
  """
  Plots a Fibonacci chart for a given dataset.

  Args:
    data: A list or array of data points.
  """

  # Calculate Fibonacci retracement levels
  fibonacci_levels = [0.0, 0.236, 0.382, 0.5, 0.618, 0.786, 1.0]

  # Find the maximum and minimum values in the data
  max_value = max(data)
  min_value = min(data)

  # Calculate the Fibonacci retracement levels in terms of price
  retracement_levels = [
    min_value + (max_value - min_value) * level
    for level in fibonacci_levels
  ]

  # Plot the original data
  plt.plot(data)

  # Plot the Fibonacci retracement levels
  for level in retracement_levels:
    plt.hlines(level, 0, len(data), linestyles='dashed')

  # Set plot title and labels
  plt.title('Fibonacci Chart')
  plt.xlabel('Time')
  plt.ylabel('Price')

  # Display the plot
  plt.show()


# Example usage:
data = [100, 102, 105, 103, 108, 110, 107, 112]

plot_fibonacci_chart(data)
```
