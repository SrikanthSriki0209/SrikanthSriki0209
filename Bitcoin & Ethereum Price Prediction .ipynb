```python
#Simple Moving Average Price Of Bitcoin & Ethereum

import pandas as pd
import matplotlib.pyplot as plt

# Load the Bitcoin and Ethereum price data
bitcoin_data = pd.read_csv("/content/Bitcoin Price 2021-2024.csv")
ethereum_data = pd.read_csv("/content/Ethreum price 2021- 2024.csv")

# Define the window size for the SMA
window_size = 50

# Calculate the SMA for Bitcoin and Ethereum prices
bitcoin_sma = bitcoin_data['Close'].rolling(window=window_size).mean()
ethereum_sma = ethereum_data['Close'].rolling(window=window_size).mean()

# Plot the SMA for both cryptocurrencies
plt.figure(figsize=(10, 6))
plt.plot(bitcoin_data['Date'], bitcoin_sma, label="Bitcoin SMA")
plt.plot(ethereum_data['Date'], ethereum_sma, label="Ethereum SMA")
plt.title('Simple Moving Average (SMA) of Bitcoin and Ethereum Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Use the SMA for price prediction
# ... (implement your prediction logic here)
```

```python
#EMA INDICATOR

import pandas as pd
from datetime import datetime

bitcoin_data = pd.read_csv("/content/Bitcoin Price 2021-2024.csv")
ethereum_data = pd.read_csv("/content/Ethreum price 2021- 2024.csv")


# Calculate the 12-day EMA for Bitcoin and Ethereum
btc_ema_12 = btc_data.ewm(span=12, adjust=False).mean()
eth_ema_12 = eth_data.ewm(span=12, adjust=False).mean()

# Calculate the 26-day EMA for Bitcoin and Ethereum
btc_ema_26 = btc_data.ewm(span=26, adjust=False).mean()
eth_ema_26 = eth_data.ewm(span=26, adjust=False).mean()

# Plot the EMA lines
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(btc_data, label="BTC Price")
plt.plot(btc_ema_12, label="BTC EMA 12")
plt.plot(btc_ema_26, label="BTC EMA 26")
plt.title("Bitcoin Price and EMA")
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(eth_data, label="ETH Price")
plt.plot(eth_ema_12, label="ETH EMA 12")
plt.plot(eth_ema_26, label="ETH EMA 26")
plt.title("Ethereum Price and EMA")
plt.legend()
plt.show()
```

```python
import pandas as pd
from datetime import datetime

bitcoin_data = pd.read_csv("/Bitcoin Price 2021-2024.csv")


# Calculate typical price
btc_data["Typical Price"] = (btc_data["Open"] + btc_data["High"] + btc_data["Low"]) / 3

# Calculate price change
btc_data["Price Change"] = btc_data["Typical Price"] - btc_data["Typical Price"].shift(1)

# Calculate positive and negative price changes
btc_data["Positive Change"] = btc_data["Price Change"].where(btc_data["Price Change"] > 0, 0)
btc_data["Negative Change"] = btc_data["Price Change"].where(btc_data["Price Change"] < 0, 0)

# Calculate average gains and losses
average_gain = btc_data["Positive Change"].rolling(14).mean()
average_loss = abs(btc_data["Negative Change"].rolling(14).mean())

# Calculate Relative Strength Index (RSI)
rsi = 20000 - (10000 / (1 + average_gain / average_loss))

# Plot the RSI
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(btc_data["Typical Price"], label="BTC Price")
plt.plot(rsi, label="RSI")
plt.title("Bitcoin Price With  Indicator RSI")
plt.legend()
plt.show()
```

```python
#  Fibonacci retracement on Bitcoin & Etherum

import numpy as np
import matplotlib.pyplot as plt

# Download the Bitcoin and Ethereum price data
btc_prices = pd.read_csv("/content/Bitcoin Price 2021-2024.csv")
eth_prices = pd.read_csv("/content/Ethreum price 2021- 2024.csv")

# Define the Fibonacci retracement levels
fibonacci_levels = [0, 0.236, 0.382, 0.5, 0.618, 0.786, 1]

# Calculate the retracement levels for Bitcoin and Ethereum
btc_retracement_levels = []
eth_retracement_levels = []

for level in fibonacci_levels:
    highest_price_btc = btc_prices["Close"].max()
    lowest_price_btc = btc_prices["Close"].min()
    btc_retracement_levels.append(highest_price_btc - level * (highest_price_btc - lowest_price_btc))

    highest_price_eth = eth_prices["Close"].max()
    lowest_price_eth = eth_prices["Close"].min()
    eth_retracement_levels.append(highest_price_eth - level * (highest_price_eth - lowest_price_eth))

# Plot the Bitcoin price and the Fibonacci retracement levels
plt.figure(figsize=(10, 6))
plt.plot(btc_prices["Close"], label="Bitcoin Price")

for level, price in zip(fibonacci_levels, btc_retracement_levels):
    plt.axhline(y=price, color="gray", linestyle="--", label=f"{level * 100:.2f}%")

plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Bitcoin Price with Fibonacci Retracement Levels")
plt.legend()

# Plot the Ethereum price and the Fibonacci retracement levels
plt.figure(figsize=(10, 6))
plt.plot(eth_prices["Close"], label="Ethereum Price")

for level, price in zip(fibonacci_levels, eth_retracement_levels):
    plt.axhline(y=price, color="gray", linestyle="--", label=f"{level * 100:.2f}%")

plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Ethereum Price with Fibonacci Retracement Levels")
plt.legend()
```
